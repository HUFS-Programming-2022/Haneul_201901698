#### list comprehension
- 포함형/지능형/축약형 리스트
- 하나의 리스트에서 그걸 포함하여 다른 리스트를 만든다.
- list + for loop
- `[A for B in C]`
    - C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
    - A : 내가 원하는 출력 형태
    - B : C의 형태
    - C : 반복 가능한 자료형 (주로 리스트)

- 조건 추가 : `[A for A' in C if ...]`

        ex: much -> m-ch
            m.........ch (하이픈의 개수도 ... 만큼 늘어나야 할 것
        - ch로 시작하면 바꾸면 안됨

```python
#2
lis = [word for word in set(text2) if word.endswith('ch')]
newlis = []
for w in lis:
    wlis = list(w) # 문자열을 리스트로 변환
    if wlis[0] == 'c' and wlis[1] == 'h':
        wlis[1] = 'h'
    else:
        wlis[1] = '-'
    for i in range(2, len(wlis)-2):
            wlis[i] = '-'
    s = ''.join(wlis) # 리스트를 문자열로 다시 변환
    newlis.append(s)
newlis.sort(key=len) # 길이 순으로 리스트 정렬 : sort(key=len)
newlis
```
